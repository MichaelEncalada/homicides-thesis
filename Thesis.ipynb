# Gráfico 1: Evolución en los años 

```python
df_homicidio["ANIO"].dtype
```

```python
# Asegurarse de que la columna 'ANIO' está en formato numérico
#df_homicidio["ANIO"] = pd.to_numeric(df_homicidio["ANIO"], errors='coerce')

# Filtrar los datos para excluir el año 2025
df_homicidio_filtrado = df_homicidio[df_homicidio["ANIO"] != 2025]

# Contar homicidios por año después de excluir 2025
homicidios_por_anio = df_homicidio_filtrado["ANIO"].value_counts().sort_index()

# Mostrar el conteo por año
print(homicidios_por_anio)
```
